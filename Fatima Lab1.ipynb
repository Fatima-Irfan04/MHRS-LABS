```python
from rasterio.plot import show
#import geotiff
#import tifffile
import matplotlib.pyplot as plt
import cv2
import numpy as np
# FOR IMAGE 1
try:
       # Read image from disk.
    img1= cv2.imread("D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif")
    
     # Get number of pixel horizontally and vertically. 
    (height, width) = img1.shape[:2]    

     # Specify the size of image along with interpolation methods.
    # cv2.INTER_AREA is used for shrinking, whereas cv2.INTER_CUBIC
    # is used for zooming.
    res1= cv2.resize(img1, (int(width / 2), int(height / 2)), interpolation=cv2.INTER_CUBIC)

    # Write image back to disk.
    cv2.imwrite('result1.tiff', res1)
 
except IOError:
    print('Error while reading files !!!')
plt.imshow(res1)
```

```python
# FOR IMAGE 2
try:
    img2= cv2.imread("D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HV-ALPSRP243410530-H2.2_UA.tif")
    (height2, width2) = img2.shape[:2]    

    res2= cv2.resize(img2, (int(width2 / 2), int(height2 / 2)), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('result2.tiff', res2)
 
except IOError:
    print('Error while reading files !!!')
plt.imshow(res2)
```

# Rotating Tiff Images

```python
#Rotating Image 1
try:
    # Read image from disk.
    img1 = cv2.imread("D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif")
 
     # Shape of image in terms of pixels.
    (rows1, cols1) = img1.shape[:2]
 
    # getRotationMatrix2D creates a matrix needed for transformation.
    # We want matrix for rotation w.r.t center to 45 degree without scaling.
    M = cv2.getRotationMatrix2D((cols1 / 2, rows1 / 2), 45, 1)
    res1 = cv2.warpAffine(img1, M, (cols1, rows1))
 
    # Write image back to disk.
    cv2.imwrite('result1.tiff', res1)
 
except IOError:
    print('Error while reading files !!!')
plt.imshow(res1)
```

```python
#Rotating Image 2
try:
    # Read image from disk.
    img2 = cv2.imread("D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HV-ALPSRP243410530-H2.2_UA.tif")
 
     # Shape of image in terms of pixels.
    (rows2, cols2) = img2.shape[:2]
 
    # getRotationMatrix2D creates a matrix needed for transformation.
    # We want matrix for rotation w.r.t center to 45 degree without scaling.
    M = cv2.getRotationMatrix2D((cols2 / 2, rows2 / 2), 45, 1)
    res2 = cv2.warpAffine(img2, M, (cols2, rows2))
 
    # Write image back to disk.
    cv2.imwrite('result2.tiff', res2)
 
except IOError:
    print('Error while reading files !!!')
plt.imshow(res2)
```

# Translating Matrix

```python
# Shifting image 1 by (100, 50).

M = np.float32([[1, 0, 100], [0, 1, 50]])
try:
    # Read image from disk.
    img1 = cv2.imread("D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif")
 
     # warpAffine does appropriate shifting given the
    # translation matrix.
    res3 = cv2.warpAffine(img1, M, (cols1, rows1))
 
    # Write image back to disk.
    cv2.imwrite('result1.tiff', res3)
 
except IOError:
    print('Error while reading files !!!')
plt.imshow(res3)
```
