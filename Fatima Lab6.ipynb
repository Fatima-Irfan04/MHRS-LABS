# NOISE REDUCTION

```python
###### NOISE REDUCTION#######

import cv2
import numpy as np
from matplotlib import pyplot as plt

image1 = cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif', 0)  
image2 = cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HV-ALPSRP243410530-H2.2_UA.tif', 0)  

blurred_image1 = cv2.GaussianBlur(image1, (5, 5), 0) 
blurred_image2 = cv2.GaussianBlur(image2, (5, 5), 0)
plt.figure(figsize=(12, 4))
plt.subplot(141), plt.imshow(image1, cmap='gray')
plt.title('Original Image 1'), plt.xticks([]), plt.yticks([])

plt.subplot(142), plt.imshow(blurred_image1, cmap='gray')
plt.title('Denoised Image 1'), plt.xticks([]), plt.yticks([])

plt.subplot(143), plt.imshow(image2, cmap='gray')
plt.title('Original Image 2'), plt.xticks([]), plt.yticks([])

plt.subplot(144), plt.imshow(blurred_image2, cmap='gray')
plt.title('Denoised Image 2'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
cv2.imwrite('denoised_image1.tif', blurred_image1)
cv2.imwrite('denoised_image2.tif', blurred_image2)
```

# Color Mapping

```python
## COLOR MAPPING #####
import numpy as np
image1 = cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif', 0)  
image2 = cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HV-ALPSRP243410530-H2.2_UA.tif', 0)  
blurred_image1 = cv2.GaussianBlur(image1, (5, 5), 0) 
blurred_image2 = cv2.GaussianBlur(image2, (5, 5), 0)

colormap_image1 = cv2.applyColorMap(blurred_image1, cv2.COLORMAP_JET)
colormap_image2 = cv2.applyColorMap(blurred_image2, cv2.COLORMAP_JET)
plt.figure(figsize=(12, 4))
plt.subplot(121),plt.imshow(cv2.cvtColor(colormap_image1, cv2.COLOR_BGR2RGB))
plt.title('Denoised Image_1 (Color-Mapped)')

plt.subplot(122),plt.imshow(cv2.cvtColor(colormap_image2, cv2.COLOR_BGR2RGB))
plt.title('Denoised Image_2 (Color-Mapped)')
plt.show()
```

# Mean Squared Error

```python
##### MEAN SQUARED ERROR #####
def mse(image1, image2):
    return np.mean((image1 - image2) ** 2)

original_image1 =cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif', 0)  
denoised_image1 = cv2.GaussianBlur(original_image1, (5, 5), 0)
mse_value = mse(original_image1, denoised_image1)
print(f"Mean Squared Error of HH image: {mse_value}")


original_image2 =cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HV-ALPSRP243410530-H2.2_UA.tif', 0)  
denoised_image2 = cv2.GaussianBlur(original_image2, (5, 5), 0)
mse_value2 = mse(original_image2, denoised_image2)
print(f"Mean Squared Error of HV image: {mse_value2}")
```

# Structural Similarity Index (SSIM)

```python
#### Structural Similarity Index (SSIM) #############
from skimage.metrics import structural_similarity as ssim
import cv2

original_image1 =cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif', 0)  
denoised_image1 = cv2.GaussianBlur(original_image1, (5, 5), 0)
ssim_value1 = ssim(original_image1, denoised_image1)
print(f"Structural Similarity Index (SSIM) of image 1: {ssim_value1}")

original_image2 =cv2.imread('D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HV-ALPSRP243410530-H2.2_UA.tif', 0)  
denoised_image2 = cv2.GaussianBlur(original_image2, (5, 5), 0)
ssim_value2 = ssim(original_image2, denoised_image2)
print(f"Structural Similarity Index (SSIM) of image 2: {ssim_value2}")
```

# FourierTransform

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original grayscale image
original_image = cv2.imread(r'D:\\ISPA\\8th Semester\\MHRS\\Data\\ALPSRP243410530-H2.2_UA\\HH-ALPSRP243410530-H2.2_UA.tif', 0)

# Load the denoised image
denoised_image = cv2.imread('denoised_image1.tif', 0)

# Ensure the images are not None
if original_image is not None and denoised_image1 is not None:
    # Compute the Fourier transform of the original image
    original_transform = np.fft.fft2(original_image)
    original_transform = np.fft.fftshift(original_transform)
    original_magnitude = np.abs(original_transform)
    original_phase = np.angle(original_transform)

    # Compute the Fourier transform of the denoised image
    denoised_transform = np.fft.fft2(denoised_image1)
    denoised_transform = np.fft.fftshift(denoised_transform)
    denoised_magnitude = np.abs(denoised_transform)
    denoised_phase = np.angle(denoised_transform)
    
     # Print amplitude (magnitude) values for both images
    print("Original Image Amplitude (Magnitude):")
    print(original_magnitude)

    print("Denoised Image Amplitude (Magnitude):")
    print(denoised_magnitude)

    # Print phase values for both images
    print("Original Image Phase:")
    print(original_phase)

    print("Denoised Image Phase:")
    print(denoised_phase)

    # Display the original image and its Fourier transform
    plt.subplot(221), plt.imshow(original_image, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(222), plt.imshow(np.log(original_magnitude), cmap='gray')
    plt.title('Original Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

    plt.subplot(223), plt.imshow(original_phase, cmap='gray')
    plt.title('Original Phase Spectrum'), plt.xticks([]), plt.yticks([])

    # Display the denoised image and its Fourier transform
    plt.subplot(224), plt.imshow(denoised_image, cmap='gray')
    plt.title('Denoised Image1'), plt.xticks([]), plt.yticks([])

    plt.show()

else:
    print("Error: Unable to load image(s). Please check the file paths.")
```
